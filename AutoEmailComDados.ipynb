# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui as pgui
import pyperclip
import time
import pandas

pgui.PAUSE = 1

# entrar no sistema
pgui.hotkey('ctrl','t')
pyperclip.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing')
pgui.hotkey('ctrl','v')
pgui.press('enter')
time.sleep(8)

# navegar no sistema
pgui.click(x=407, y=261, clicks=2)
time.sleep(8)

# baixar base de vendas
pgui.click(x=428, y=318)
pgui.click(x=1712, y=162)
pgui.click(x=1507, y=563)
time.sleep(8)
pgui.press('enter')
time.sleep(6)
pgui.hotkey('ctrl','w')
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# importar base pro python
tabela = pandas.read_excel(r'C:\Users\jhrma\Downloads\Vendas - Dez.xlsx')
display(tabela)

# calcular faturamento e quantidade de produtos
faturamento = tabela['Valor Final'].sum()
qtprodutos = tabela['Quantidade'].sum()
print(faturamento)
print(qtprodutos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# entrar no email
pgui.hotkey('ctrl','t')
pyperclip.copy('https://mail.google.com/mail/u/0/#inbox')
pgui.hotkey('ctrl','v')
pgui.press('enter')
time.sleep(6)

# navegar no email e preencher campos
pgui.click(x=99, y=169)
pyperclip.copy('jhrmatos@hotmail.com')
pgui.hotkey('ctrl','v')
pgui.press('tab')
pyperclip.copy('Aula: Intensivão de Python')
pgui.hotkey('ctrl','v')
pgui.press('tab')

texto = f"""
Olá, João!

Parabéns, você está programando!

O faturamento do mês foi de R${faturamento:,.2f}
A quantidade de produtos vendidos foi de {qtprodutos:,}

Att,
Você mesmo
"""

pyperclip.copy(texto)
pgui.hotkey('ctrl','v')
pgui.hotkey('ctrl','enter')
time.sleep(5)
pgui.hotkey('ctrl','w')
```
